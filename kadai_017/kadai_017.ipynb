```python
class Human:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print("大人である。")
        else:
            print("大人でない。")

a1 = Human("tanaka", 35)
a2 = Human("sato", 18)
a3 = Human("suzuki", 32)

users = [a1, a2, a3]

for i in users:
    i.check_adult()
```
